```python
import math as mt
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl

g = 9.8066 # На высоте 0м над уровнем моря, т.к. корабли, вроде по морю ходят и в высоту <<км
dt = 0.01
vo = 910
y0 = 30
class Kinn:
    def __init__(self, x,y,vx,vy,y0):
        """Обозначние переменных, которые будут использоваться для описания движения снаряда"""
        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy

        self.trajectory_x = []
        self.trajectory_y = []
    def advance(self):
        """Описание физики полёта и запись промежуточных результатов"""
        self.trajectory_x.append(self.x)
        self.trajectory_y.append(self.y)

        self.x += self.vx * dt
        self.y += y0 + self.vy * dt
        self.vy -= g*dt
class React(Kinn):
    def __init__(self,x,y,m,dm,ta,deg,u,vx,vy):
        """Обозначние переменных, которые будут использоваться для описания движения ракеты"""
        self.deg = deg
        self.x = x
        self.y = y
        self.m = m
        self.dm = dm
        self.vx = vx
        self.vy = vy
        self.deg = deg
        self.u = u
        self.ta = ta
        self.trajectory_x = []
        self.trajectory_y = []
    def advance (self):
        """Описание физики полёта и запись промежуточных результатов"""
        self.trajectory_x.append(self.x)
        self.trajectory_y.append(self.y)
        if t<self.ta: # Т.е. используется кинетика ракетного тела до тех пор, пока есть топливо (ta – вермя работы ракетного двигателя)
            self.vx += ((i*self.u)/(self.m-i))*np.cos(self.deg)
            self.vy += ((i*self.u)/(self.m-i))*np.sin(self.deg)-g*dt
            self.x += self.vx*dt
            self.y += self.vy*dt
            self.m -= i
        else:
            self.x += self.vx * dt
            self.y += self.vy * dt
            self.vy -= g*dt
k = Kinn(0,0,788.08,455,27) # При угле возвышения в 30 градусов, 27
r = React(0.0,0.0,3.85,0.815,0.75,np.pi/4,1600,0.0,0.0)
i = r.dm/(r.ta/dt) # Это Unit mass – уменьшение массы ракеты за 1 заданный период dt
bodies = [r,k]

for t in np.r_[0:20:dt]:
    for b in bodies:
        b.advance()
        plt.plot(b.trajectory_x, b.trajectory_y) # Траектория снаряда выглядит как прямая, но, если поставить хотя бы 100, она выглядит лучше
plt.show()
```
